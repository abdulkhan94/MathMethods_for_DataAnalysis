# Importing required packages

```python
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import statistics as stat
```

# Setting up data

```python
diabetes = pd.read_csv('./diabetes.csv')
pd.set_option('precision', 2) #Setting dataframe precision
pd.set_option('display.max_rows', 10000) #Setting max rows to display
```

# Calculating mean, median, standard deviation of all age values 
#### We will compare this mean, median & standard deviation against the results from the Central Limit Theorem

```python
print(f'The mean for the entire sample is {stat.mean(diabetes.Age):.2f}, the median is {stat.median(diabetes.Age):.2f} and the standard deviation is {stat.pstdev(diabetes.Age):.2f}')
```

### Converting dataframe column Age to numpy array for calculation

```python
ages = diabetes.Age.to_numpy()
```

# -------------------------------------------------------------------------------------------------

# Mean

#### Calculating mean for Sample Size = 15, Sample Count = 100

```python
mean_list15_100 = []  # Will store all the means for a 100 runs to plot

for _ in range(100): 
    age_list15_100 = np.random.choice(ages, size =15, replace= False)
    mean_list15_100.append(stat.mean(age_list15_100)) # Appends the mean of the 15 samples selected each run
    
offset_15_100 = stat.mean(mean_list15_100) - stat.mean(diabetes.Age) # Calculates offset
```

#### Distribution of Sample Size = 15, Sample Count = 100

```python
sns.set_style('darkgrid')
sns.set_context('talk')
axes = sns.distplot(mean_list15_100, bins= 15, color = 'blue')
axes.set_xlabel('Age')
axes.set_title(f'Sample Size: {len(age_list15_100)} & Sample Count: {len(mean_list15_100)}');
```

```python
mean_list15_1000 = []  # Will store all the means for a 1000 runs to plot

for _ in range(1000): 
    age_list15_1000 = np.random.choice(ages, size =15, replace= False)
    mean_list15_1000.append(stat.mean(age_list15_1000)) # Stores the mean
    
offset_15_1000 = stat.mean(mean_list15_1000) - stat.mean(diabetes.Age)
```
