# Importing required packages

```python
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import statistics as stat
```

# Setting up data

```python
diabetes = pd.read_csv('./diabetes.csv')
pd.set_option('precision', 2) #Setting dataframe precision
pd.set_option('display.max_rows', 10000) #Setting max rows to display
```

# Calculating mean, median, standard deviation of all age values 
#### We will compare this mean, median & standard deviation against the results from the Central Limit Theorem

```python
print(f'The mean for the entire sample is {stat.mean(diabetes.Age):.2f}, the median is {stat.median(diabetes.Age):.2f} and the standard deviation is {stat.pstdev(diabetes.Age):.2f}')
```

### Converting dataframe column Age to numpy array for calculation

```python
ages = diabetes.Age.to_numpy()
```

# Analysis:


For our analysis to estimate central theorem of statistics, we have chosen a dataset "diabetes" which has the information relating to the insulin impairement spread across the age groups from 21 - 81.

            As a first step, we sought to calculate 3 population parameters - mean, median, std deviation for our analysis.
            The calculated parameters on the total dataset stood at:
                                    Mean = 33.24
                                    Median = 29.00
                                    Standard deviation = 11.75
           
In order to aid our analysis of central limit theorem, random samples were drawn to analyse the parameters in consideration. The method choosen to draw samples was using a Numpy function - random.choice() with varying sample sizes and sample counts.

## Calculated Mean for Randomly drawn Samples::
     
          1.  We have first started with random sample size of 15 and a count of 1000, which resulted in calculated mean of 32.00 
          2.  Second random sample size was still limited to 15 while the sample count was increased to 2000, which again resulted in a mean 
              of 32.00
          
Since both exeriments listed above resulted with sample size 15 though sample counts are at 1000 & 2000, the resulted mean was observed 
at 32.00 which is 1.24 lesser than the mean of total dataset measured at 33.24

In order to further our analysis, we have now increased the sample size to 30 and chose the sample counts at 100 and 1000 respectively.

          3.  For the drawn sample with size 30 and count 100, the mean stood at the same 32.00
          4.  The sample with size 30 and count 1000 has recorded mean of 32.00
          
Based on our above simulations to analyse the mean, we could observe that the mean from all samples with varying amounts of sizes and counts stood at 32.00 which is an offset of 1.24 from the overall mean of the dataset.

Few observations that we could identify are:
 
           - Calculated Means stood in the Normal range(distributed normally)
           - As we increased the size of samples, it indeed smoothened the curve as can be seen above in the bar
             charts.

```python
median_list05_100 = []  # Will store all the medians for a 100 runs to plot

for _ in range(100): 
    median_age_list05_100 = np.random.choice(ages, size =5, replace= False)
    median_list05_100.append(stat.median(median_age_list05_100)) # Appends the median of the 30 samples selected each run
    
median_offset_05_100 = stat.median(median_list05_100) - stat.median(diabetes.Age) # Calculates offset

print(f'As you can see in the above distribution with a sample size of {len(median_age_list05_100)} & a sample count of {len(median_list05_100)} the median of the distribution was {stat.median(median_list05_100)} which is an offset of {median_offset_05_100:.2f} from the calculated median of the entire sample')
```

## Calculated Standard deviation for Randomly drawn Samples1:

For the analysis of third population parameter standard deviation, we have considered varying sampling sizes and counts to analyse the parameter standard deviation.

         1. For the sample size 15 & the sample count 100, the standard deviation was calculated to be 2.21
         2. For the sample size 15, this time with an increased samples to 1000, the standard deviation was measured as 2.42

In the above two samplings, we found that while the sample size remained at 15, the increased number of samples from 100 to 1000, actually increased the deviation from 2.21 to 2.42 which was confusing.

Perplexed by the increase in deviation, we had decided to conduct further simulations to gain better insights on the parameter standard deviation. 

This time we have choosen to start with a smaller size to start with since the earlier experiments with size 15 and increased counts actually increased standard deviation. 
The objective of these additional simulations was to: 

        a. Explore the corelation between size and counts in relation to 'standard deviation'.
        b. Understand the impact of increased size on deviation.
        
Additional simulation results:

   - We have started with a lesser sample size than earlier simulations and,

         3. For the sample size 05 and the count 50, the standard deviation was observed as 3.80
               
            Observation: a decreased sample size incomparision to earlier size of 15 actually increased the standard deviation.
            
   - Based on the outcome of experiment 3, we have increased the sample size to 30 as opposed to our simulations in 1 and 2,
   
         4. For the same size 30 and the count 100, the standard deviation was measured 1.76
         
            Observation: an increased sample size actually reduced the standard deviation.
            
   - Based on the outcome of experiment 4, we decided to continue with sample size 30 while increasing the sample count to 1000 and,
         
         5. For the sample size 30 and the count of 1000, the standard deviation was measured to be 1.73 
         
            Observation: an increased sample count has minimal impact on standard deviation as long as the sample size is  of a good quality.
            
   - Based on the outcome of experiment 5, we have decided further test our observation in experiment 4 by furthering the sample size to 50 
 
         6. For the sample size 50 and the count 500, the standard deviation was calculated at 1.31
         
             Observation: We have again noticed that with an increase in sample size we have a reduced standard deviation from 1.73 to 1.31.
             
   - As a final experiment, we decided to test the standard deviation by increasing the size to 100(which is a larger sample) and continuing 
     with the earlier sample count of 500.
     
         7. For the sample size of 100 and the sample count of 500, the standard deviation was measured at 0.92
         

As a final conclusion, we could understand that an increased sample size actually reduces the standard deviation or margin of error as was seen in our final experiment in comparision to other experiments though this was a difficult parameter to measure and understand.
