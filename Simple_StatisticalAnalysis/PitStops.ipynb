### Merging two data frames to create relevant visualizations & analysis

```python
drivers_pitstops = pd.merge(drivers_top, pitstops_top, how='inner', on = 'driverId')
drivers_pitstops.columns = ['driverID', 'driverRef', 'Number', 'Code' , 'Forename' , 'Surname', 'DOB', 'Nationality', 'URL', 'raceId', 'Stop', 'Lap', 'Time', 'Duration', 'Milliseconds']
```
